```python
import pandas as pd
from statsmodels.stats.inter_rater import fleiss_kappa, aggregate_raters
import numpy as np
```

```python
df = pd.read_csv("annotation.csv")
df = df.drop(['Unnamed: 0','profession'], axis=1)
```

```python
df
```

```python
pivot_df = df.pivot(index='filename', columns='annotator', values=['annotation'])
pivot_df.columns = [f'{b}' for a, b in pivot_df.columns]
print(pivot_df.isnull().sum())
final_df = pivot_df.dropna()
```

```python
agg_df = aggregate_raters(np.array(final_df))
# https://en.wikipedia.org/wiki/Fleiss%27_kappa
fleiss_kappa(agg_df[0])
```
